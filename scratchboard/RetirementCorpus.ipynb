# Retirement Corpus Calculator

## Initialization

```python
import numpy as np
import matplotlib.pyplot as plt
```

## User Defined Input Variables

```python
years = 35
principal = 10000000
monthlyExpense = 41711
inflation = 6
returnRate = 10
```

## Calculation

```python
balance = principal
balArr = []
gainsArr = []
expArr = []
mexpArr = []
yArr = range(years)
for i in  yArr:
    totalExpense = round(monthlyExpense * 12, 2)
    balance = round(balance - totalExpense,2)
    gains = round(balance * returnRate/100.0,2)
    balance = round(balance + gains,2)
    balArr.append(balance)
    gainsArr.append(gains)
    expArr.append(totalExpense)
    mexpArr.append(round(monthlyExpense))
    monthlyExpense = round(monthlyExpense * (1 + inflation/100.0),2)
```

## Data Table Print

```python
print("Year\tMonthlyExpense\t\t\t\tTotalExpense\t\t\t\tGains\t\t\t\tBalance")
for i in  yArr:
    print(f"{i}\t{mexpArr[i]}\t\t\t\t{expArr[i]}\t\t\t\t{gainsArr[i]}\t\t\t\t{balArr[i]}")
```

```python
plt.title("Balance over Time")
plt.xlabel("Years")
plt.ylabel("Rs")
plt.plot(yArr, np.divide(balArr,10), color="blue", label ="Balance")
plt.plot(yArr, gainsArr, color="green",label = "Gains")
plt.plot(yArr, expArr, color="red", label = "Expenses")
plt.legend()
plt.plot(yArr,np.zeros(years), color = "black")
```
